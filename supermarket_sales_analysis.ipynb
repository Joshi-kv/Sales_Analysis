```python
#finding which location has highest and lowest sales.
location_list=data.groupby('Location')
location=[x for x,y in location_list]
```

```python
location
```

```python
import matplotlib.pyplot as plt
data.groupby('Location').sum()['Total']
```

```python
plt.bar(location,data.groupby('Location').sum()['Total'])
```

```python
plt.pie(data.groupby('Location').sum()['Total'],labels=location,autopct='%1.1f%%')
```

```python
#which location has more male and female customer
gender_sales=data.groupby(['Gender','Location']).count()['Invoice ID']
gender_sales.unstack(level=0)
```

```python
gender_sales.unstack(level=0).plot(kind='bar')
```

```python
#which day of month has more sales
#adding a new column day by extracting day from date column
data['Day']=pd.to_datetime(data['Date']).dt.day
```

```python
data.head()
```

```python
day_sales=data.groupby('Day').sum()['Total']
```

```python
day_sales.plot()
```

```python
#which branch has more and less members
members=data.groupby(['Member','Location']).count()['Invoice ID']
members.unstack(level=0)
```

```python
members.unstack(level=0).plot(kind='bar')
```
